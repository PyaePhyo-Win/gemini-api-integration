```python
import os
from dotenv import load_dotenv
from google import genai
from google.genai import types
from IPython.display import display, Markdown, HTML

# 1. Load API Key from .env
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")

# 2. Initialize the Client
client = genai.Client(api_key=api_key)
```

```python
# 3. Start a Chat Session
# This session will automatically maintain message history for us.
chat = client.chats.create(model="gemini-2.5-flash")

def ask_chat(prompt):
    """
    Sends a message in the active chat session and displays the response.
    """
    try:
        response = chat.send_message(prompt)
        display(Markdown(f"**You:** {prompt}"))
        display(Markdown(f"**Gemini:** {response.text}"))
        print("-" * 30)
    except Exception as e:
        print(f"Error: {e}")
```

```python
# Turn 1
ask_chat("My name is Pyae Phyo. What's yours?")
```

```python
# Turn 2
ask_chat("Do you remember my name?")
```

```python
# 4. Accessing History
# You can inspect the current conversation history like this:
for message in chat._curated_history:
    role = message.role
    text = message.parts[0].text
    print(f"[{role.upper()}]: {text[:50]}...")
```
