# Response Streaming with Gemini

This notebook demonstrates how to stream responses from Gemini in real-time, providing a more interactive experience for multi-turn conversations.

```python
import os
from dotenv import load_dotenv
from google import genai
from google.genai import types
from IPython.display import display, Markdown, clear_output

# 1. Load API Key from .env
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")

# 2. Initialize the Client
client = genai.Client(api_key=api_key)
```

## Multi-turn Conversation with Streaming

To enable streaming in a chat session, we use the `send_message_stream` method. This allows us to process and display chunks of the response as they are generated by the model.

```python
# 3. Start a Chat Session
chat = client.chats.create(model="gemini-2.5-flash")

def ask_chat_stream(prompt):
    """
    Sends a message in the active chat session and streams the response.
    """
    display(Markdown(f"**You:** {prompt}"))
    
    print("Gemini: ", end="", flush=True)
    
    # Initialize an empty string to accumulate the response
    full_response = ""
    
    # Create a display handle to update the Markdown content
    display_handle = display(Markdown("*"), display_id=True)
    
    try:
        # Use send_message_stream for real-time output
        for chunk in chat.send_message_stream(prompt):
            if chunk.text:
                full_response += chunk.text
                # Update the displayed Markdown with the accumulated text
                display_handle.update(Markdown(full_response))
        
        print("\n" + "-" * 30)
    except Exception as e:
        print(f"\nError: {e}")
```

### Interactive Conversation

Try running the cells below to see the streaming in action!

```python
# Turn 1
ask_chat_stream("Explain the concept of quantum entanglement in simple terms.")
```

```python
# Turn 2
ask_chat_stream("That was great! Now, how does it relate to quantum computing?")
```

### Verifying History

Even with streaming, the `chat` object maintains the full conversation history. You can verify this by printing the history.

```python
for message in chat._curated_history:
    role = "User" if message.role == "user" else "Gemini"
    print(f"{role}: {message.parts[0].text[:50]}...")
```
