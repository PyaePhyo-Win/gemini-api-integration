# 💬 Interactive Gemini Chat Exercise

This notebook implements a continuous chat loop. 
- Type your message and press **Enter**.
- Type **'exit'**, **'quit'**, or **stop the cell** to end the session.

```python
import os
from dotenv import load_dotenv
from google import genai
from google.genai import types
from IPython.display import display, Markdown, HTML, clear_output
import time

# 1. Setup Client
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")
client = genai.Client(api_key=api_key)

def display_usage(usage):
    """Utility to display a beautiful token dashboard"""
    prompt_t = usage.prompt_token_count
    cand_t = usage.candidates_token_count
    think_t = getattr(usage, 'thoughts_token_count', 0) 
    total_t = usage.total_token_count

    dashboard_html = f"""
    <div style="font-family: sans-serif; border: 1.5px solid #333; border-radius: 10px; 
                padding: 10px; background-color: #1a1a1a; color: #e0e0e0; 
                max-width: 350px; margin: 10px 0; font-size: 0.8em;">
        <div style="color: #4ec9b0; font-weight: bold; border-bottom: 1px solid #333; margin-bottom: 5px;">
            📊 Session Stats
        </div>
        <div style="display: grid; grid-template-columns: 1fr 1fr;">
            <div style="color: #9cdcfe;">Input: {prompt_t}</div>
            <div style="color: #ce9178; text-align: right;">Thinking: {think_t}</div>
            <div style="color: #b5cea8;">Output: {cand_t}</div>
            <div style="color: #ffffff; font-weight: bold; text-align: right;">Total: {total_t}</div>
        </div>
    </div>
    """
    display(HTML(dashboard_html))
```

```python
# 2. Start Chat Session
chat = client.chats.create(model="gemini-2.5-flash")

print("🚀 Chat Session Started! (Type 'exit' to quit)")
print("-" * 50)

try:
    while True:
        # Get user input
        user_input = input("You: ")
        
        # Exit conditions
        if user_input.lower() in ['exit', 'quit', 'stop']:
            print("\n👋 Chat session ended by user.")
            break
            
        if not user_input.strip():
            continue

        # Send to Gemini
        response = chat.send_message(
            user_input, 
            config=types.GenerateContentConfig(
                thinking_config=types.ThinkingConfig(include_thoughts=True)
            )
        )
        
        # Display Response
        display(Markdown(f"**Gemini:** {response.text}"))
        
        # Show Token Usage
        display_usage(response.usage_metadata)
        print("-" * 50)

except KeyboardInterrupt:
    print("\n\n🛑 Session interrupted (KeyboardInterrupt).")
except Exception as e:
    print(f"\n❌ An error occurred: {e}")
finally:
    print("✅ Goodbye!")
```
