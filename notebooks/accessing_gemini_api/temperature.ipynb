# 🌡️ Temperature Control in Gemini

This notebook demonstrates how to adjust the **Temperature** parameter.
- **Low Temperature (e.g., 0.0):** More deterministic, focused, and predictable. Great for facts or logic.
- **High Temperature (e.g., 1.0+):** More creative, diverse, and unexpected. Great for brainstorming or stories.

```python
import os
from dotenv import load_dotenv
from google import genai
from google.genai import types
from IPython.display import display, Markdown

# Setup
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")
client = genai.Client(api_key=api_key)
```

## 1. Normal (Single-Turn) with Temperature
This function sends a one-off request with a custom temperature.

```python
def ask_single_with_temp(prompt, temp=0.7):
    """Sends a single-turn request with a specific temperature."""
    try:
        response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=prompt,
            config=types.GenerateContentConfig(
                temperature=temp
            )
        )
        display(Markdown(f"### Temperature: {temp}"))
        display(Markdown(response.text))
        print("-" * 50)
    except Exception as e:
        print(f"Error: {e}")
```

```python
prompt = "Write a one-sentence story about a lonely robot on Mars."

print("Deterministic Response:")
ask_single_with_temp(prompt, temp=0.0)

print("Creative Response:")
ask_single_with_temp(prompt, temp=1.5)
```

## 2. Multi-Turn (Chat) with Temperature
This function initializes a chat session where the temperature is set for the entire conversation.

```python
def start_chat_with_temp(temp=0.7):
    """Creates a chat session with a fixed temperature."""
    return client.chats.create(
        model="gemini-2.5-flash",
        config=types.GenerateContentConfig(
            temperature=temp
        )
    )

def ask_chat(chat_session, prompt):
    """Sends a message to the chat session."""
    response = chat_session.send_message(prompt)
    display(Markdown(f"**Gemini:** {response.text}"))
```

```python
# Create a high-temperature creative chat session
creative_chat = start_chat_with_temp(temp=1.2)

print("Creative Chat Turn 1:")
ask_chat(creative_chat, "Let's invent a new type of fruit. Describe it.")

print("\nCreative Chat Turn 2:")
ask_chat(creative_chat, "What does it taste like when you cook it?")
```
