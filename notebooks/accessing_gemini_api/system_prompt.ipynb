# 🧠 Multi-Turn System Prompt Exercise: Math Tutor

This notebook demonstrates how to maintain a **multi-turn conversation** while using a **System Prompt** to define the model's behavior.

```python
import os
from dotenv import load_dotenv
from google import genai
from google.genai import types
from IPython.display import display, Markdown

# Setup
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")
client = genai.Client(api_key=api_key)

# 1. Define the Reusable System Prompt
MATH_TUTOR_PROMPT = """
You are a helpful and encouraging math tutor for middle school students. 
Your goal is to guide the student to find the answer themselves. 
1. NEVER provide the final answer immediately.
2. Ask leading questions to help them think through the steps.
3. If they are stuck, provide a small hint or explain the underlying mathematical concept.
4. Use a friendly, patient, and encouraging tone.
"""
```

```python
def create_tutor_chat(use_tutor_mode=True):
    """
    Initializes a chat session. If use_tutor_mode is True, 
    it applies the system prompt for the entire session.
    """
    system_instruction = MATH_TUTOR_PROMPT if use_tutor_mode else None
    
    return client.chats.create(
        model="gemini-2.5-flash",
        config=types.GenerateContentConfig(
            system_instruction=system_instruction
        )
    )

def ask_tutor(chat_session, prompt):
    """
    Sends a message to an existing chat session and displays the response.
    """
    try:
        response = chat_session.send_message(prompt)
        display(Markdown(f"**You:** {prompt}"))
        display(Markdown(f"**Gemini:** {response.text}"))
        print("-" * 50)
    except Exception as e:
        print(f"Error: {e}")
```

## Multi-Turn Session with System Prompt
Observe how the tutor stays in character across multiple messages.

```python
# Start a session with tutor mode ENABLED
tutor_session = create_tutor_chat(use_tutor_mode=True)

# Turn 1
ask_tutor(tutor_session, "How do I solve 2x + 5 = 13?")
```

```python
# Turn 2
ask_tutor(tutor_session, "I subtracted 5 from both sides, now I have 2x = 8. What's next?")
```

```python
# Turn 3
ask_tutor(tutor_session, "Oh! So x must be 4, right?")
```

## Comparison: Standard Mode (No System Prompt)
In standard mode, the model will likely just give you the answer immediately.

```python
# Start a session with tutor mode DISABLED
standard_session = create_tutor_chat(use_tutor_mode=False)

ask_tutor(standard_session, "How do I solve 2x + 5 = 13?")
```
